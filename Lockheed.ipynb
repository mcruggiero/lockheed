A "noun chunk" is a simplification of a sentence. 

### Spacy data collection

```python
nchunks = []
sentence_array = []
html_percent = []

for string in df.text.values:
    nlp_analysis = nlp(string)
    chunker = [str(x.lemma_).lower().strip() for x in nlp_analysis.noun_chunks if str(x).find('http:') == -1]
    
    # We don't want to collect urls
    nchunks.append(chunker)
    
    # It will be useful to collect if a tweet has a url
    html_count = sum([1 if str(x).find('http:') > -1 else 0 for x in nlp_analysis.noun_chunks])
    
    # It is possible that the len of the chunk list could = 0, add small correction
    html_percent.append(html_count / len(string))
    
    sentence_array.append(nlp_analysis.vector)
```

### Noun Chunk Creation

```python
noun_corpus = sorted(list(set(x for sublist in nchunks for x in sublist)))
```

```python
noun_corpus[2000:2003]
```

It is not necessary to manually program the corpus collection for nouns, but I thought it was informative to show a programatic option.

#### Train test split

#### Tokenize Text based on training

```python
tokenizer = get_tokenizer('basic_english')
counter = Counter()

# Note that we are building counter vocab on training set
for (label, line) in train_iter:
    counter.update(tokenizer(line))

# This is an important decison node for the n-gram size
vocab = Vocab(counter, min_freq=4)
```

```python
# A lambda function is just a little one liner, called with text_pipeline(x)
text_pipeline = lambda x: [vocab[token] for token in tokenizer(x)]
label_pipeline = lambda x: int(x)
```

```python
text_pipeline('OMG here is the an example')
```

#### Batch builder / dataloader

```python
def collate_batch(batch):
    label_list, text_list, offsets = [], [], [0]
    for (_label, _text) in batch:
        label_list.append(label_pipeline(_label))
        
        # Turns text into tensors
        processed_text = torch.tensor(text_pipeline(_text), dtype=torch.int64)
        text_list.append(processed_text)
        offsets.append(processed_text.size(0))
    label_list = torch.tensor(label_list, dtype=torch.int64)
    offsets = torch.tensor(offsets[:-1]).cumsum(dim=0)
    text_list = torch.cat(text_list)
    return label_list.to(device), text_list.to(device), offsets.to(device)
```

```python
dataloader = DataLoader(train_iter, batch_size=8, shuffle=True, collate_fn=collate_batch)
```

#### NN design

```python
class TextClassificationModel(nn.Module):

    def __init__(self, vocab_size, embed_dim, num_class):
        super(TextClassificationModel, self).__init__()
        self.embedding = nn.EmbeddingBag(vocab_size, embed_dim, sparse=True)
        self.fc = nn.Linear(embed_dim, num_class)
        self.init_weights()

    def init_weights(self):
        initrange = 0.5
        self.embedding.weight.data.uniform_(-initrange, initrange)
        self.fc.weight.data.uniform_(-initrange, initrange)
        self.fc.bias.data.zero_()

    def forward(self, text, offsets):
        embedded = self.embedding(text, offsets)
        return self.fc(embedded)
```

#### Initial model

```python
num_class = len(set([label for (label, text) in train_iter]))
vocab_size = len(vocab)
emsize = 64
model = TextClassificationModel(vocab_size, emsize, num_class).to(device)
```

#### Train/Evaluate Functions

```python
def train(dataloader):
    model.train()
    total_acc, total_count = 0, 0
    log_interval = 500
    start_time = time.time()

    for idx, (label, text, offsets) in enumerate(dataloader):
        optimizer.zero_grad()
        predited_label = model(text, offsets)
        loss = criterion(predited_label, label)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 0.1)
        optimizer.step()
        total_acc += (predited_label.argmax(1) == label).sum().item()
        total_count += label.size(0)
        if idx % log_interval == 0 and idx > 0:
            elapsed = time.time() - start_time
            print('| epoch {:3d} | {:5d}/{:5d} batches '
                  '| accuracy {:8.3f}'.format(epoch, idx, len(dataloader),
                                              total_acc/total_count))
            total_acc, total_count = 0, 0
            start_time = time.time()

def evaluate(dataloader):
    model.eval()
    total_acc, total_count = 0, 0

    with torch.no_grad():
        for idx, (label, text, offsets) in enumerate(dataloader):
            predited_label = model(text, offsets)
            loss = criterion(predited_label, label)
            total_acc += (predited_label.argmax(1) == label).sum().item()
            total_count += label.size(0)
    return total_acc/total_count
```

#### Tokenized Model

```python
# Hyperparameters
EPOCHS = 15 # epoch
LR = 5  # learning rate
BATCH_SIZE = 64 # batch size for training

criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=LR)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, 1.0, gamma=0.1)
total_accu = None
train_dataset = list(train_iter)
test_dataset = list(test_iter)
num_train = int(len(train_dataset) * 0.95)
split_train_, split_valid_ = \
    random_split(train_dataset, [num_train, len(train_dataset) - num_train])

train_dataloader = DataLoader(split_train_, batch_size=BATCH_SIZE,
                              shuffle=True, collate_fn=collate_batch)
valid_dataloader = DataLoader(split_valid_, batch_size=BATCH_SIZE,
                              shuffle=True, collate_fn=collate_batch)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE,
                             shuffle=True, collate_fn=collate_batch)

for epoch in range(1, EPOCHS + 1):
    epoch_start_time = time.time()
    train(train_dataloader)
    accu_val = evaluate(valid_dataloader)
    if total_accu is not None and total_accu > accu_val:
      scheduler.step()
    else:
       total_accu = accu_val
    print('-' * 59)
    print('| end of epoch {:3d} | time: {:5.2f}s | '
          'valid accuracy {:8.3f} '.format(epoch,
                                           time.time() - epoch_start_time,
                                           accu_val))
    print('-' * 59)
```

```python
class Model(nn.Module):
    def __init__(self, n_input_features):
        super(Model, self).__init__()
        self.input_linear = nn.Linear(n_input_features, 2 * n_input_features)
        self.hidden_linear = nn.Linear(2 * n_input_features, n_input_features)
        self.output_linear = nn.Linear(n_input_features, 1)
        
    def forward(self, x):
        # Input layer
        x = self.input_linear(x)
        x = F.dropout(x, p=0.75)
        x = x.add(np.random.uniform(-.45, .45))
        x = F.leaky_relu(x)
        
        # hidden layer
        x = self.hidden_linear(x)
        x = F.dropout(x, p=0.75)
        x = x.add(np.random.uniform(-.45, .45))
        x = F.leaky_relu(x)
        
        # Output layer
        x = self.output_linear(x)        
        y_predicted = torch.sigmoid(x)
        
        return y_predicted
    
model = Model(n_features)
```

### Type I/II errors

```python
def error_class(trouble):
    if trouble["target"] == 0 and trouble["prediction"] == 0:
        return "TN"
    elif trouble["target"] == 1 and trouble["prediction"] == 1:
        return "TP"
    elif trouble["target"] == 1 and trouble["prediction"] == 0:
        return "FN"
    else:
        return "FP"
```

```python
trouble["error_type"] = trouble.apply(error_class, axis = 1)
```
